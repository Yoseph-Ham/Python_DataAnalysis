```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                   'B' : ['B0', 'B1', 'B2', 'B3'],
                   'C' : ['C0', 'C1', 'C2', 'C3'],
                   'D' : ['D0', 'D1', 'D2', 'D3']},
                  index = [0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                   'B' : ['B4', 'B5', 'B6', 'B7'],
                   'C' : ['C4', 'C5', 'C6', 'C7'],
                   'D' : ['D4', 'D5', 'D6', 'D7']},
                  index = [4, 5, 6, 7])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                   'B' : ['B8', 'B9', 'B10', 'B11'],
                   'C' : ['C8', 'C9', 'C10', 'C11'],
                   'D' : ['D8', 'D9', 'D10', 'D11']},
                  index = [8, 9, 10, 11])
```

```python
df1
```

```python
df2
```

```python
df3
```

# concat

## 단순 병합

```python
result = pd.concat([df1, df2, df3])
result
```

## key 지정 병합

```python
result = pd.concat([df1, df2, df3], keys = ['x', 'y', 'z'])
result
```

```python
result.index # 다중 인덱스 확인
```

## 다중 인덱스 조회

```python
result.index.get_level_values(0)
```

```python
result.index.get_level_values(1)
```

## concat은 index 기준으로 병합

```python
df4 = pd.DataFrame({'B' : ['B2', 'B3', 'B6', 'B7'],
                   'C' : ['C2', 'C3', 'C6', 'C7'],
                   'D' : ['D2', 'D3', 'D6', 'D7']},
                  index = [2, 3, 6, 7])
result = pd.concat([df1, df4], axis=1)
```

```python
result
```

## 공통되지 않은 index의 데이터 버리기 join = 'inner'

```python
result = pd.concat([df1, df4], axis=1, join='inner')
result
```

## df1의 인덱스에 맞추어 병합하기

```python
result = pd.concat([df1, df4], axis=1).reindex(df1.index)
result
```

## 인덱스 무시하고 병합하기

```python
result = pd.concat([df1, df4], axis=1, ignore_index = True)
result
```

## Merge

```python
left = pd.DataFrame({
                    'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'key': ['K0', 'K4', 'K2', 'K3']})

right = pd.DataFrame({'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3'],
                     'key': ['K0', 'K1', 'K2', 'K3']})
```

```python
left
```

```python
right
```

```python
# 공통된 key에서만 병합
pd.merge(left, right, on = 'key')
```

```python
pd.merge(left, right, how = 'left', on = 'key') # left 기준 병합
```

```python
pd.merge(left, right, how = 'right', on = 'key') # right 기준 병합
```

```python
pd.merge(left, right, how = 'outer', on = 'key') # 합집합
```

```python
pd.merge(left, right, how = 'inner', on = 'key') # 교집합
```

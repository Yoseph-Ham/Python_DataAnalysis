### seaborn을 import 할 때는 matplotlib도 같이 import 해야함

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

x = np.linspace(0, 14, 100)
y1 = np.sin(x)
y2 = 2*np.sin(x+0.5)
y3 = 3*np.sin(x+1)
y4 = 4*np.sin(x+4)

plt.figure(figsize = (10, 6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

### Seaborn의 whitegrid 스타일

```python
sns.set_style('whitegrid')

plt.figure(figsize = (10, 6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

### Seaborn의 연습 데이터셋

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('whitegrid')
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

### X축은 요일, Y축은 전체 금액으로 하는 박스플롯

```python
plt.figure(figsize = (8, 6))
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
plt.show()
```

### hue 옵션으로 흡연, 비흡연 여부 확인하기

```python
plt.figure(figsize = (8, 6))
sns.boxplot(x = 'day', y = 'total_bill', hue = 'smoker', data = tips, palette = "Set3")
plt.show()
```

### Seaborn의 lmplot

### 총 지불 금액과 팁의 관계
### 데이터를 scatter처럼 그리고 직선으로 regression한 그림도 같이 그려주고 유효범위를 ci로 잡음

```python
sns.set_style("darkgrid") #  darkgrid  스타일
sns.lmplot(x = "total_bill", y = "tip", data = tips, height = 7)
plt.show()
```

### hue 옵션을 사용하여 흡연, 비흡연을 구별

```python
sns.set_style("darkgrid") #  darkgrid  스타일
sns.lmplot(x = "total_bill", y = "tip", hue = "smoker", data = tips, height = 7, palette = "Set1")
plt.show()
```

## Seaborn의 연습데이터 flights

```python
flights = sns.load_dataset("flights")
flights.head(5)
```

```python
flights = flights.pivot("month", "year", "passengers")
flights.head()
```

### Seaborn의 heatmap
### 데이터의 경향을 설명하기 좋음

```python
plt.figure(figsize = (10, 8))
sns.heatmap(flights, annot = True, fmt = "d")
plt.show()
```

```python
sns.set(style = "ticks")
iris = sns.load_dataset("iris")
iris.head(10)
```

### Seaborn의 pairplot
### 꽃잎, 꽃받침의 너비, 폭으로 종을 구분

```python
sns.pairplot(iris, hue = "species")
plt.show()
```

###  history 시장가(시가, 종가, 고가, 저가), 거래량 데이터 가져오기

history()메서드에 사용 가능한 매개변수는 다음 과 같습니다.

기간 : 다운로드할 데이터 기간(기간 매개변수 사용 또는 시작 및 종료 사용) 유효 기간: 1d, 5d, 1mo, 3mo, 6mo, 1y, 2y, 5y, 10y, ytd, max

간격 : 데이터 간격(일중 데이터는 지난 60일 동안 연장할 수 없음) 유효한 간격: 1m, 2m, 5m, 15m, 30m, 60m, 90m, 1h, 1d, 5d, 1wk, 1mo, 3mo

start : 마침표를 사용하지 않는 경우 - 시작 날짜 문자열(YYYY-MM-DD) 또는 날짜/시간을 다운로드합니다.

end : 마침표를 사용하지 않는 경우 - 종료 날짜 문자열(YYYY-MM-DD) 또는 날짜/시간을 다운로드합니다.

prepost : 결과에 프리장 및 애프터장 데이터를 포함합니까? (기본값은 False)

auto_adjust : 모든 OHLC를 자동으로 조정하시겠습니까? (기본값은 True)

action : 주식 배당 및 주식 분할 이벤트를 다운로드하시겠습니까? (기본값은 True)

```python
msft.history(period = 'max')
```

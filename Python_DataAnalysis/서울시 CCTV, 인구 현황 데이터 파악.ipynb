```python
import pandas as pd
import numpy as np
```

```python
CCTV_Seoul = pd.read_csv('C:/Users/has91/anaconda3/DataScience/data/01. CCTV_in_Seoul.csv')
CCTV_Seoul.rename(columns = {CCTV_Seoul.columns[0] : '구별'}, inplace = True)
CCTV_Seoul.head()
```

```python
pop_Seoul = pd.read_excel('C:/Users/has91/anaconda3/DataScience/data/01. population_in_Seoul.xls',
                         header = 2,
                         usecols = 'B, D, G, J, N')
pop_Seoul.head()
```

```python
pop_Seoul.rename(columns = {pop_Seoul.columns[0] : "구별",
                pop_Seoul.columns[1] : "인구수",
                pop_Seoul.columns[2] : "한국인",
                pop_Seoul.columns[3] : "외국인",
                pop_Seoul.columns[4] : "고령자"},
                inplace = True)
pop_Seoul.head()
```

## CCTV 최근 증가율 계산

### 2013년 대비 CCTV 증가율

```python
CCTV_Seoul['최근증가율'] = (CCTV_Seoul['2016년'] + CCTV_Seoul['2015년'] + CCTV_Seoul['2014년']) / CCTV_Seoul['2013년도 이전'] * 100
CCTV_Seoul.sort_values(by='최근증가율', ascending=False).head(5)
```

### 종로구, 도봉구, 마포구, 노원구, 강동구 -  최근 3년간 CCTV가 많이 증가함

## 외국인비율, 고령자 비율 계산

```python
pop_Seoul['외국인 비율'] = pop_Seoul['외국인'] / pop_Seoul['인구수'] * 100
pop_Seoul['고령자 비율'] = pop_Seoul['고령자'] / pop_Seoul['인구수'] * 100
pop_Seoul.head()
```

## '구별' 열로 병합

```python
data_result = pd.merge(CCTV_Seoul, pop_Seoul, on = '구별')
data_result.head()
```

## 의미없는 컬럼 제거

```python
del data_result['2013년도 이전']
del data_result['2014년']
del data_result['2015년']
del data_result['2016년']
data_result.head()
```

## index 설정 - set_index

```python
data_result.set_index('구별', inplace = True)
data_result.head()
```

# 상관관계 계수 (np.corrcoef)

## 상관계수의 절대값이 클 수록 두 데이터는 관계가 있다.
### 0.1 이하면 거의 무시, 0.3이하면 약한 상관관계 0.7 이하면 뚜렷한 상관관계

## CCTV 갯수와 고령자 비율, 외국인 비율, 인구수와의 상관관계

```python
np.corrcoef(data_result['고령자 비율'], data_result['소계'])
```

```python
np.corrcoef(data_result['외국인 비율'], data_result['소계'])
```

```python
np.corrcoef(data_result['인구수'], data_result['소계'])
```
